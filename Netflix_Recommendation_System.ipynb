# Reading dataset

```python
import pandas as pd

movies = pd.read_csv('/content/drive/MyDrive/movies.csv')
movies.head(1000)
```

# Dropping not required columns

```python
required_columns =['title','original_title', 'tagline', 'keywords', 'overview', 'genres', 'cast', 'director']
movies = movies[required_columns]
```

```python
movies.isna().sum()
```

```python
#replacing the missing values with single space character
movies.fillna(' ', inplace=True)
movies.isna().sum()
```

```python
#The iloc method is used to select rows and columns by integer position.
movies.iloc[1]
```

# Create movie CONTENT

```python
#these lines of code are removing spaces from the values of specific columns in the 'movies' dataframe,
#making the data consistent and easier to work with during later stages of the program.
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director'] = movies['director'].apply(lambda x: x.replace(" ","") )
```

## Converting List to Strings

```python
# genres, keywords, cast which are list of strings are converted into comma separated strings

movies['genres'] = movies['genres'].apply(lambda x: ','.join(map(str, x)))
movies['keywords'] = movies['keywords'].apply(lambda x: ','.join(map(str, x)))
movies['cast'] = movies['cast'].apply(lambda x: ','.join(map(str, x)))
```

```python
movies['content'] = movies['title'] + ' ' + movies['overview'] + ' ' + movies['keywords'] + ' ' + movies['cast'] + ' ' + movies ['director']
movies['content']
```

# Natural Language Processing

```python
# TF-IDF Vectorizer is a method used to transform text documents into a numerical format
# TF-IDF stands for "Term Frequency-Inverse Document Frequency"

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000)
movie_vectors = vectorizer.fit_transform(movies['content'].values) 
```

# Cosine Similarity
The cosine similarity is a metric that measures the similarity between two vectors in a high-dimensional space. In the context of this code, it measures the similarity between pairs of movie feature vectors based on the words they contain and how often those words appear in each movie's plot summary, cast, director, and keywords. Cosine similarity ranges from -1 to 1, with values closer to 1 indicating that the movies are more similar and values closer to -1 indicating that the movies are less similar. The resulting similarity matrix can be used to recommend similar movies to users based on their viewing history or to cluster similar movies together for further analysis.

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(movie_vectors)
```

```python
similarity_df = pd.DataFrame(similarity)
similarity_df
```

# Recommending Movies

```python
def recommend(movie):
    # find movie index from dataset
    movies_index = movies[movies['title'] == movie].index[0]
    
    # finding cosine similarities of movie
    distances = similarity[movies_index]
    
    # sorting cosine similarities
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```
